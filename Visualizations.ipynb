# Visualizations for the Tests' results

## Instance Recovery Test

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':14})


# Input of type [Main-Shadow, Main-App, Shadow-Main, APP-LB]
recovery_time = np.array([[8.082, 6.072, 11.038, 13.133], [12.031, 6.05, 12.047, 16.034], 
                          [6.026, 4.058, 5.025, 5.025], [4.025, 4.04, 3.024, 4.028]])
mean_time = np.mean(recovery_time, 1)
std_time = np.std(recovery_time, 1)
print("mean:",mean_time)
print("std:",std_time)

test_time = np.array([30.163, 20.219, 31.133, 38.219])
print("Mean duration of test:", np.mean(test_time))
print("Std of test duration:", np.std(test_time))

plt.figure(figsize=(8,5))
plt.bar(range(4), mean_time, yerr=std_time, width=0.7)
plt.xlabel("Component Type")
plt.ylabel("Recovery time (min)")
plt.xticks(range(4), ("Main - Shadow", "Main - AppOrchestrator", "Shadow - Main",
                      "AppOrchestrator - LoadBalancer"), rotation=20)
```

## Performance Test

```python
import numpy as np
time = np.array([[0.902, 0.588, 0.595, 0.604], [2.604, 2.543, 2.5, 2.453], 
                 [4.933, 4.973, 5.199, 6.203], [63.054, 11.927, 12.642, 11.884]])
mean_time = np.mean(time, 1)
std_time = np.std(time, 1)
print("mean:",mean_time)
print("std:",std_time)

test_time = np.array([71.493, 20.031, 20.936, 21.142])
print("Mean duration of test:", np.mean(test_time))
print("Std of test duration:", np.std(test_time))

plt.figure(figsize=(8,5))
plt.ylabel("Time (min) for completion of requests")
plt.xlabel("Number of submitted requests")
plt.plot(range(4), mean_time, 'o-', color='b')
plt.fill_between(range(4), mean_time-std_time, mean_time+std_time, alpha=0.1, color='b')
plt.xticks(range(4), ("1", "10", "20", "50"))

plt.figure(figsize=(8,5))
plt.bar(range(4), mean_time, yerr=std_time, width=0.7)
plt.ylabel("Time (min) for completion of requests")
plt.xlabel("Number of submitted requests")
plt.xticks(range(4), ("1", "10", "20", "50"))
```
