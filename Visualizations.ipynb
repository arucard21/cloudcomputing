# Visualizations for the Tests' results

## Instance Recovery Test

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':14})


# Input of type [Main-Shadow, Main-App, Shadow-Main, APP-LB]
recovery_time = np.array([11.039, 13.032,7.045, 18.04])
test_duration = 1
print("Mean:", np.mean(recovery_time))

plt.figure(figsize=(8,5))
plt.bar(range(4), recovery_time, yerr=np.ones((4,1)), width=0.7)
plt.xlabel("Component Type")
plt.ylabel("Recovery time (min)")
plt.xticks(range(4), ("Main - Shadow", "Main - AppOrchestrator", "Shadow - Main",
                      "AppOrchestrator - LoadBalancer"), rotation=20)
```

## Performance Test

```python
time = np.array([0.574, 2.496, 4.868, 11.78])
std_time = np.ones(4)

test_duration = 1

plt.figure(figsize=(8,5))
plt.ylabel("Time (min) for completion of requests")
plt.xlabel("Number of submitted requests")
plt.plot(range(4), time, 'o-', color='b')
plt.fill_between(range(4), time-std_time, time+std_time, alpha=0.1, color='b')
plt.xticks(range(4), ("1", "10", "20", "50"))

plt.figure(figsize=(8,5))
plt.bar(range(4), time, yerr=std_time, width=0.7)
plt.ylabel("Time (min) for completion of requests")
plt.xlabel("Number of submitted requests")
plt.xticks(range(4), ("1", "10", "20", "50"))
```
